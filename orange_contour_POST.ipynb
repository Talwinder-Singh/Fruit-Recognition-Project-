```python
import cv2
import numpy as np
import json
import requests
import matplotlib.pyplot as plt

import math
from flask import Flask, jsonify
import numpy as np
# import argparse
import time
import cv2
from flask import request
import os
import matplotlib.pyplot as plt
from os import listdir
from os.path import isfile, join
import numpy
import cv2

# %matplotlib inline
app = Flask(__name__)


@app.route('/post', methods=["POST"])
def contour():
    file = request.get_json()
    mypath = file["path"]
    onlyfiles = [ f for f in listdir(mypath) if isfile(join(mypath,f)) ]
    images = numpy.empty(len(onlyfiles), dtype=object)
    for n in range(0, len(onlyfiles)):
        images[n] = cv2.imread( join(mypath,onlyfiles[n]) )
    #plt.imshow(cv2.cvtColor(images[n], cv2.COLOR_BGR2RGB))
        img_HSV = cv2.cvtColor(images[n], cv2.COLOR_BGR2HSV)
        HSV_mask = cv2.inRange(img_HSV, (0, 80, 20), (25, 255, 255))
        target = cv2.bitwise_and(images[n],images[n], mask=HSV_mask)
        plt.imshow(cv2.cvtColor(target, cv2.COLOR_BGR2RGB))

        contours,hierarchy  = cv2.findContours(cv2.cvtColor(target, cv2.COLOR_BGR2GRAY), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        cv2.drawContours(images[n], contours, -1, (0,255,0), 3)

        #contours,hierarchy  = cv2.findContours(cv2.cvtColor(target, cv2.COLOR_BGR2GRAY), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
        for contour in contours:
            area = cv2.contourArea(contour)
            #print(area)
            bounding_boxes = []
            threshold = 75
            ar=[]
            imgwithbox = images[n].copy()
        for c in contours:
            area = cv2.contourArea(c)
            if area >threshold:
                [x,y,w,h]=cv2.boundingRect(c)
                bounding_boxes.append([x,y,w,h])
                #print([x,y,w,h])
                area_rect=w*h
                #print(area_rect)
                ar.append(area_rect)
                cv2.rectangle(imgwithbox,(x,y),(x+w,y+h),(0,255,0),2)

        #print(len(bounding_boxes))
        #print(ar)
        total_area=sum(ar)
        if(len(bounding_boxes)>0):#print(total_area)
            avg_rect= total_area/len(bounding_boxes)
            #print(avg_rect)

            q=np.asarray(ar)
            number_of_oranges_list=[]
            #print(q)
            no_of_oranges = q/avg_rect
            #print(no_of_oranges)
            rounded_value=np.around(no_of_oranges,0)
            #print(rounded_value)
            total_fruits=sum(rounded_value)
            #print(total_fruits)
            number_of_oranges=[]
            #print(no_of_oranges)
            for count in enumerate(no_of_oranges):
        #     print(index)
                if count[1]<0.3:
                    no_of_oranges[count[0]]=0
                elif count[1]>0.3 and count[1]<1.5:
                    no_of_oranges[count[0]]=1
                elif count[1]>1.5 and count[1]<2.5:
                    no_of_oranges[count[0]]=2
                elif count[1]>2.5 and count[1]<3.5:
                    no_of_oranges[count[0]]=3
            no_of_oranges
            fruits=sum(no_of_oranges )
            #print(int(fruits))
        output = images[n].copy()
    # .putText(output, "WELCOME TO PROJECT FRUIT", (10, 25),
        cv2.putText(output,f" Number  of  fruits : {fruits}", (10, 25), 

        cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        cv2.imshow("Text", output)        #cv2.imshow("Text", output)
        #cv2.imwrite("/home/talwinder/Desktop/xyz2.jpg",output)
        #n=np.array

        #x=images[n]
    #---- 'x' contains the input image for every iteration

    #-----
    #-- Here is where you write the code to process your images--
    #-----

        #xy = str(images[(n)])
    #-----'xy' is the new extension you add for every iterated image

        #cv2.imwrite('/home/talwinder/Desktop/orange_result/' + xy,output)
        cv2.imwrite("/home/talwinder/Desktop/orange-result/"+str(n)+".jpg",output)
        
        #print({"fruits":str(fruits)})
        cv2.waitKey(0)
        cv2.destroyAllWindows()
    
    return "Success"


if __name__ == '__main__':
    app.run(port=8008)
```
